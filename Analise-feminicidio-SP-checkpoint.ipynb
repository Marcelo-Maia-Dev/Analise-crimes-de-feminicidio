# Análise dos Dados de crimes de feminicidios - São Paulo

**SSP - TRANSPARÊNCIA**

Desde 1995, a Secretaria da Segurança Pública divulga de forma pioneira no Brasil os indicadores criminais trimestralmente no Diário oficial.

Em decorrência da evolução da tecnologia e o amplo uso da internet, a partir de abril de 2011 esses dados passaram a ser divulgados em seu site mensalmente por região, município e unidade policial.

Além desses dados, são apresentados estudos como o Perfil de Homicídio e o Perfil de Roubos além da produtividade policial.

**Neste notebook, irei analisar os dados referentes à crimes de feminicio no estado de São Paulo, e ver quais insights podem ser extraídos a partir de dados brutos.**

## Obtenção dos Dados

Todos os dados usados aqui foram obtidos a partir do site [Segurança publica - transparência](http://www.ssp.sp.gov.br/transparenciassp/Consulta.aspx).

Para esta análise exploratória inicial, será baixado apenas o seguinte arquivo:

```python
# importar os pacotes necessarios
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# importar o arquivo Feminicidio_2015_2022.xlsx para um Dataframe
df = pd.read_excel(r"C:\Users\Maia\Desktop\MENTORIA\python\.ipynb_checkpoints\Feminicidio_2015_2022.xlsx")
display(df)
```

### Número total de vitimas de feminicidio em São Paulo

```python
# Verificando o número total de vítimas
Num_total = df["Nº DE VÍT HD"].sum()
print(Num_total)
```

### Prosseguindo e identificando a quantidade de entradas que o conjunto de dados possui e verificando os tipos de cada coluna


```python
# identificar o volume de dados do DataFrame
print("Entradas:\t {}".format(df.shape[0]))
print("Variáveis:\t {}\n".format(df.shape[1]))

# verificar as 5 primeiras entradas do dataset
display(df.dtypes)
```

 ### Para identificar a distribuição das variáveis, irei plotar o histograma.

```python
# plotar o histograma das variáveis numéricas
df.hist(bins=15, figsize=(15,10));
```

### Quais municipios mais cometem esse tipo de crime?

```python
# Total de crimes por municipio 
Num_municipio = df[["MUNICIPIO_CIRCUNSCRICAO","Nº DE VÍT HD"]].groupby("MUNICIPIO_CIRCUNSCRICAO").sum()
display(Num_municipio)
```
